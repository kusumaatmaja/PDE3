## Plotting a 1D Function

In this section we will cover the basics of plotting using matplotlib in Python.

Before we start, we need to create a figure to add lines to.
We can do this by calling the [`plt.figure()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) function.

For a simple line plot, we can use the [`plt.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) function. This function takes in two main arguments, the x and y values of the data to be plotted.

Finally, we can show the figure using the [`plt.show()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html) function.

```python
# First, we create some data to plot
# Find out more about using numpy in the 'Using Numpy Arrays.ipynb' notebook
x = np.linspace(0, 3*np.pi, 100)
y = np.sin(x/2)

# Now we can plot the data
plt.figure()
plt.plot(x, y)
plt.show()
```

We have made a basic plot, however it is missing important information such as axis labels and a title. A legend is also useful when plotting multiple lines on the same plot.
These can be added using the following functions:
* [`plt.xlabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html)
* [`plt.ylabel()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html)
* [`plt.title()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html)
* [`plt.legend()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

## Plotting 2D Heatmaps

Since this course is about solving PDEs (which have solutions with more than one variable), we often also need to plot 2D heatmaps.
These take an array of 2D data and plot it as a heatmap.

To do this, we can use the [`plt.pcolor()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pcolor.html) function. This function takes in the x and y values of the data, as well as the data itself.
The plot will be coloured according to the values in the data array.

To make sense of the colours in the plot, we can add a colour bar using the [`plt.colorbar()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.colorbar.html) function.

As before, it is important to add axis labels and a title to the plot.

```python
# First, we need to generate some data to plot:

xs = np.linspace(0, 2 * np.pi, 100)
ys = np.linspace(0, 2 * np.pi, 100)

# When we use a 2D plot, we need to convert our 1D x and y arrays to 2D arrays.
# You don't need to worry too much about how this works as the grids you use will often be set up for you.
X, Y = np.meshgrid(xs, ys)
Z = np.sin(X) * np.cos(Y)

# Now we can plot the data
plt.figure()
plt.pcolor(X, Y, Z)
plt.colorbar()

# Add a title and labels
plt.title('sin(x) * cos(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Sometimes, it may also be useful to plot multiple plots on the same figure. This can be done using the [`plt.subplot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html) function. You can find out more about this function [here](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html).

Below, we remake the figure given as an example above, improving the formatting of the plot.

```python
# Use the same data as before:

xs = np.linspace(0, 2 * np.pi, 100)
ys = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(xs, ys)
Z = np.sin(X) * np.cos(Y)

# Plot the data

fig, ax = plt.subplots(1, 1, figsize=(6.5,6))
c = ax.pcolor(X, Y, Z, cmap='viridis')
cax = fig.colorbar(c, ax=ax)

# Add a title and labels
plt.title(r'$z=\sin(x)\cos(y)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
cax.set_label(r'$z$')

#Set the tick marks

plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
plt.yticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
cax.set_ticks([-1,-0.5, 0, 0.5,1])

plt.show()
```
